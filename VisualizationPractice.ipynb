```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

# BAR CHART

```python
df=pd.read_csv("bar_chart_data.csv")
```

```python
df
```

```python
#figure size
plt.figure(figsize=(9,6))

bar= plt.bar(x=df["Brand"], #x-axis
        height=df["Cars Listings"], #y-axis
        color= "darkblue") #colour of bars

# Add labels to bars
plt.bar_label(bar)

plt.xticks(rotation=45) #rotation of x-axis labels
plt.title("Cars Listings by Brand",fontsize=20,fontweight="bold") #title of figure
plt.xlabel("Brands", fontsize=15) #label for x-axis
plt.ylabel("Cars Listing", fontsize=15) #label for y-axis
plt.savefig('cars_bar_plot.png', dpi=300, bbox_inches='tight') #to save the figure with complete info
plt.show() #final command to print the visualization
```

# PIE CHART

# STACKED AREA CHARTS


```python
df_stacked_area=pd.read_csv("stacked_area_chart_data.csv")
df_stacked_area
```

```python
plt.figure(figsize=(11,6))

sns.set_style("white")

colors=["#011683","#7e2987","#ef2026"]
labels=["Gas","Petrol","Diesel"]
plt.stackplot(df_stacked_area["Year"],
              df_stacked_area["Gas"],
              df_stacked_area["Petrol"],
              df_stacked_area["Diesel"],
             colors=colors,
             edgecolor="none")

plt.xticks(df_stacked_area["Year"], rotation=45)

plt.legend(labels=labels, loc="upper left")
plt.ylabel("Number of Cars",fontsize=13)

plt.title("Popularity of Engine Fuel Types (1982-2016)",fontsize=15,fontweight="bold")

sns.despine() #to remove borders
plt.show()
```

# SCATTER PLOTS

```python
df_scatter=pd.read_csv("scatter_data.csv")
```

```python
df_scatter
```

```python
plt.figure(figsize=(12,8))
sns.set()
scatter=plt.scatter(df_scatter["Area (ft.)"],
            df_scatter["Price"],
            alpha=0.6, #transperancy to handle overplotting probelm
            c= df_scatter["Building Type"], #color the plots with respect to third variable 
            cmap="viridis") # mako/viridis

plt.legend(*scatter.legend_elements(),
           loc="upper left",
            title="Building Type")
plt.title("Relationship between Area and Prices with respect to Building Types in California Real Estate",fontsize=15,fontweight="bold")
plt.xlabel("Area (ft.)", fontweight="bold")
plt.ylabel("Price", fontweight="bold")

plt.show()
```

```python
#using seaborn

plt.figure(figsize=(12,8))

sns.scatterplot(x=df_scatter["Price"],
                y=df_scatter["Area (ft.)"],
               hue=df_scatter["Building Type"],
               palette=["black","white","blue","red","pink"],
               s=100) #size of each point

plt.title("Relationship between Area and Prices with respect to Building Types in California Real Estate",fontsize=15,fontweight="bold")
plt.xlabel("Area (ft.)", fontweight="bold")
plt.ylabel("Price", fontweight="bold")

plt.show()
```

# Regression plot

```python
df_reg=pd.read_csv("scatter_plot_ii.csv")
df_reg
```

```python
# plt.figure(figsize=(9,6))
sns.set(rc={"figure.figsize":(9,6)}) #using seaborn
sns.regplot(x="Budget",
            y="Sales",
            data=df_reg,
           # color="gray"
           scatter_kws={"color":"k"},
           line_kws={"color":"red"})

plt.title("Regression plot of Budget and Sales",fontsize=15,fontweight="bold")

plt.show()
```

# LINE-BAR CHART

```python
df_line_bar=pd.read_csv("bar_line_chart_data.csv")
```

```python
df_line_bar
```

```python
from matplotlib.ticker import PercentFormatter
```

```python
sns.set_style("white")
fig, ax=plt.subplots(figsize=(9,6)) #mutlpile visulaizations==multiple axes

ax.bar(df_line_bar["Year"],
       df_line_bar["Participants"],
      color="k")

ax.set_xlabel("Year", fontweight="bold")
ax.set_ylabel("Number of Partcipants",
             fontweight="bold")

ax1=ax.twinx() #as they share same x-axis

ax1.set_ylim(0,1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax=1.0)) #to get percentages on y-axis

ax1.plot(df_line_bar["Year"],
         df_line_bar["Python Users"],
        color="Red",
        marker="D")

ax1.set_ylabel("Percentage of Python Users",
              color="Red",
              fontweight="bold")

plt.title("KD Nuggets Uers Survey (2012-2019)",fontsize=15,fontweight="bold")

plt.show()
```
